## 加载数据

## 1-1 数据修改



### 1- 数据修改｜列名



将原 `df` 列名 `Unnamed: 2`、`Unnamed: 3`、`Unnamed: 4` 修改为 `金牌数`、`银牌数`、`铜牌数`

```python
df.rename(columns={'Unnamed: 2': '金牌数',
                   'Unnamed: 3': '银牌数',
                   'Unnamed: 4': '铜牌数'}, inplace=True)
```

```python
df
```

### 2 - 数据修改｜行索引



将第一列（排名）设置为索引

```python
df.rename(index=df['排名'], inplace=True)
```

```python
df
```

### 3 - 数据修改｜修改索引名



修改索引名为 金牌排名

```python
df.index.name = '金牌排名'
df
```

### 4 - 数据修改｜修改值



将 ROC（第一列第五行）修改为 俄奥委会

```python
df.iloc[4, 1] = '俄奥委会'
df
```

### 5 - 数据修改｜替换值（单值）



将金牌数列的数字 `0` 替换为 `无`

```python
df['金牌数'] = df['金牌数'].replace(0, '无')
df
```

### 6 - <font color = '#FB8E00'>数据修改</font>｜替换值（多值）



同时替换

- 将 `无` 替换为 缺失值
- 将 0 替换为 `None`

注意：缺失值的 Nan 该怎么生成？

```python
import numpy as np

df.replace('无', np.nan, inplace=True)
df.replace(0, None, inplace=True)
df
```

### 7 - 数据查看

查看各列数据类型

```python
df.dtypes
```

### 8 - 数据修改｜修改类型

将 `金牌数` 列类型修改为 `int`

```python
# Ensure the '金牌数' column is of type int
df['金牌数'] = df['金牌数'].replace(np.nan, 0)
df['金牌数'] = df['金牌数'].astype(int)
df['金牌数'].dtypes
```

### 9 - 数据增加｜新增列（固定值）



**重新加载数据** 并 **新增一列** 比赛地点，值为 `东京`


```python
df['比赛地点'] = '东京'
df
```

### 10 - 数据增加｜新增列（计算值）



**新增一列** 金银牌总数列，值为该国家金银牌总数

```python
df['金银牌总数'] = df['金牌数'] + df['银牌数']
df
```

### 11 -数据增加｜新增列（比较值）

新增一列 最多奖牌数量 列，值为该国 `金、银、铜` 牌数量中最多的一个奖牌数量

例如美国银牌最多，则为41，中国为38

```python
df['最多奖牌数量'] = df[['金牌数', '银牌数', '铜牌数']].max(axis=1)
df
```

### 12 - 数据增加｜新增列（判断值）

新增一列 金牌大于30

如果一个国家的金牌数大于 30 则值为 是，反之为 否

```python
df['金牌大于30'] = df['金牌数'].apply(lambda x: '是' if x > 30 else '否')
df
```

### 13 - 数据增加｜增加多列

新增两列，分别是

- 金铜牌总数（金牌数+铜牌数）
- 银铜牌总数（银牌数+铜牌数）

```python
df['金铜牌总数'] = df[['金牌数', '铜牌数']].sum(axis=1)
df['银铜牌总数'] = df[['银牌数', '铜牌数']].sum(axis=1)

df
```

### 14 - 数据增加｜新增列（引用变量）

新增一列金牌占比，为各国金牌数除以总金牌数（gold_sum）

```python
df['金牌占比'] = df['金牌数'].div(df['金牌数'].sum())
df
```

### 15 - 数据增加｜新增行（末尾追加）



在 df 末尾追加一行，内容为 0,1,2,3... 一直到 `df` 的列长度

```python
new_row = list(range(df.shape[1]))
new_row = pd.DataFrame([new_row],columns=df.columns)
# df._append(pd.Series(new_row,index = df.columns),ignore_index=True)
df = pd.concat([df,new_row],ignore_index=True)
df
```

### 16 -数据增加｜新增行（指定位置）



在第 2 行新增一行数据，即美国和中国之间。

数据内容同上一题

```python
df = pd.concat([df.loc[:0],new_row,df.loc[1:]],ignore_index=True)
df
```

### 17 - 数据删除｜删除行



删除 `df` 第一行

```python
df = df.drop(df.index[0])
df
```

### 18 - 数据删除｜删除行（条件）
金牌赛>18的国家

```python
df =  df.drop(df.index[df['金牌数']>18])
df
```

### 19 - 数据删除｜删除列



删除刚刚新增的 比赛地点 列

```python
df = df.drop('比赛地点',axis = 1)
df
```

### 20 - 数据删除｜删除列（按列号）



删除 `df` 的 `7、8、9、10` 列

```python
df.drop(df.columns[7:11],axis = 1)
```
